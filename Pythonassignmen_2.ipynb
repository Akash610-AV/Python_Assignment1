```python
#37.Merge back the lookups with the main file (simulate master lookup join).

# Temporarily reset the index of df to make OrderID a column for merging
df_temp = df.reset_index()

# Drop the auxiliary 'index' column created by reset_index() from previous operations
# This column is a leftover from when df was previously reset and then OrderID was set as index
if 'index' in df_temp.columns:
    df_temp = df_temp.drop(columns=['index'])

# Create a proper customer master lookup (unique by CustomerID)
# This ensures that for each CustomerID, we have a single, canonical CustomerName
customer_master_df = df_temp[['CustomerID', 'CustomerName']].drop_duplicates(subset=['CustomerID'])

# Create a proper product-category master lookup (unique by Product and Category)
product_category_master_df = df_temp[['Product', 'Category']].drop_duplicates(subset=['Product', 'Category'])

# Perform the first merge with customer_master_df
# 'suffixes' are used to distinguish columns that exist in both dataframes (e.g., CustomerName)
df_merged = pd.merge(df_temp, customer_master_df, on='CustomerID', how='left', suffixes=('_main', '_cust_lookup'))

# Perform the second merge with product_category_master_df
df_merged = pd.merge(df_merged, product_category_master_df, on=['Product', 'Category'], how='left', suffixes=('_main', '_prod_lookup'))

# Update the main df and set OrderID back as its index
df = df_merged.set_index('OrderID')

print("DataFrame after merging lookups (first 5 rows):")
display(df.head())
```

```python
# Make a working copy of the DataFrame and reset its index to ensure OrderID is a column.
df_working = df.reset_index().copy()

# Ensure 'OrderDate' is datetime type.
df_working['OrderDate'] = pd.to_datetime(df_working['OrderDate'], errors='coerce')

# Drop rows with NaT in OrderDate as rolling operations require valid dates.
df_working_cleaned = df_working.dropna(subset=['OrderDate']).copy()

# Sort the DataFrame by 'City' and 'OrderDate' to ensure correct rolling window calculation.
df_working_cleaned = df_working_cleaned.sort_values(by=['City', 'OrderDate'])

# Temporarily set 'OrderDate' as the index of this DataFrame for rolling calculations.
# This allows the use of time-based window (e.g., '7D').
df_indexed_for_rolling = df_working_cleaned.set_index('OrderDate')

# Calculate the 7-day rolling average of 'TotalAmount' per city.
# The result will be a Series with a MultiIndex (City, OrderDate).
rolling_avg_series = df_indexed_for_rolling.groupby('City')['TotalAmount'].rolling('7D', min_periods=1).mean()

# Convert the MultiIndex Series back to a DataFrame and rename the column.
rolling_avg_result_df = rolling_avg_series.reset_index()
rolling_avg_result_df = rolling_avg_result_df.rename(columns={'TotalAmount': '7DayRollingAvg_TotalAmount'})

# Merge the calculated rolling averages back to the original `df_working_cleaned`
# using 'City' and 'OrderDate' to align the results with the correct rows.
df_working_cleaned = pd.merge(df_working_cleaned, rolling_avg_result_df,
                              on=['City', 'OrderDate'], how='left')

# Finally, merge the new '7DayRollingAvg_TotalAmount' column back to the main DataFrame `df`
# using 'OrderID', which is its current index.
df = pd.merge(df, df_working_cleaned[['OrderID', '7DayRollingAvg_TotalAmount']],
              on='OrderID', how='left')

print("First 10 rows with '7DayRollingAvg_TotalAmount' column:")
display(df[['City', 'OrderDate', 'TotalAmount', '7DayRollingAvg_TotalAmount']].head(10))
```

```python
#79. Normalize all text columns (lowercase, strip spaces, remove special chars if any).
import re

# Identify text columns (object dtype)
text_columns = df.select_dtypes(include=['object']).columns

for col in text_columns:
    # Convert to string, then lowercase
    df[col] = df[col].astype(str).str.lower()
    # Strip leading/trailing spaces
    df[col] = df[col].str.strip()
    # Remove special characters, keeping only alphanumeric and spaces
    # Using a regex to replace anything not a letter, number, or space
    df[col] = df[col].apply(lambda x: re.sub(r'[^a-zA-Z0-9\s]', '', x))

print("First 5 rows of normalized text columns (CustomerName, Product, Category, City, Region):")
display(df[['CustomerName_main', 'Product', 'Category', 'City', 'Region']].head())
```

```python
#80. Filter all products that contain the substring 'top' (case insensitive).
# Filter products that contain the substring 'top' (case insensitive)
products_with_top = df[df['Product'].str.contains('top', case=False, na=False)]

print("Products containing 'top' (case insensitive) (first 10 rows):")
display(products_with_top.head(10))
```

```python
#81.Use pivot_table() to summarize TotalAmount by Region and Category.
pivot_table_region_category = df.pivot_table(index='Region', columns='Category', values='TotalAmount', aggfunc='sum')

print("Pivot table of TotalAmount by Region and Category:")
display(pivot_table_region_category)
```

```python
#82.Create a column to show each order's percentage of the total monthly revenue.
# Ensure 'OrderDate' is datetime and extract 'OrderYearMonth'
df['OrderDate'] = pd.to_datetime(df['OrderDate'], errors='coerce')
df['OrderYearMonth'] = df['OrderDate'].dt.to_period('M')

# Calculate total monthly revenue
monthly_revenue = df.groupby('OrderYearMonth')['TotalAmount'].sum().reset_index()
monthly_revenue.rename(columns={'TotalAmount': 'MonthlyTotalRevenue'}, inplace=True)

# Merge monthly revenue back to the main DataFrame
df = pd.merge(df, monthly_revenue, on='OrderYearMonth', how='left')

# Calculate each order's percentage of the total monthly revenue
df['PercentageOfMonthlyRevenue'] = (df['TotalAmount'] / df['MonthlyTotalRevenue']) * 100

print("First 10 rows with 'TotalAmount', 'MonthlyTotalRevenue', and 'PercentageOfMonthlyRevenue' columns:")
display(df[['OrderDate', 'OrderYearMonth', 'TotalAmount', 'MonthlyTotalRevenue', 'PercentageOfMonthlyRevenue']].head(10))
```

```python
#83.Use .transform() to normalize TotalAmount within each Region.
# Calculate normalized TotalAmount within each Region using .transform()
df['NormalizedTotalAmount'] = df.groupby('Region')['TotalAmount'].transform(lambda x: (x - x.mean()) / x.std())

print("First 10 rows with Region, TotalAmount, and NormalizedTotalAmount columns:")
display(df[['Region', 'TotalAmount', 'NormalizedTotalAmount']].head(10))
```

```python
#84. Perform one-hot encoding of the Region column.
# Perform one-hot encoding on the 'Region' column
df_encoded = pd.get_dummies(df, columns=['Region'], prefix='Region')

print("DataFrame after one-hot encoding the 'Region' column (first 5 rows):")
display(df_encoded.head())
```

```python
#86.Create a customer lookup DataFrame with unique CustomerID, CustomerName, and total orders. Save it.
# Create a customer lookup DataFrame
customer_lookup_df = df.groupby(['CustomerID', 'CustomerName_main'])['OrderID'].nunique().reset_index()
customer_lookup_df.rename(columns={'CustomerName_main': 'CustomerName', 'OrderID': 'TotalOrders'}, inplace=True)

print("Customer Lookup DataFrame with Total Orders (first 10 rows):")
display(customer_lookup_df.head(10))
```

```python
#87.Create a product lookup DataFrame with unique Product, Category, and average Price.Save it.
# Create a product lookup DataFrame
product_lookup_df = df.groupby(['Product', 'Category'])['Price'].mean().reset_index()
product_lookup_df.rename(columns={'Price': 'AveragePrice'}, inplace=True)

print("Product Lookup DataFrame with Average Price (first 10 rows):")
display(product_lookup_df.head(10))
```

```python
#88.Merge the original Sales_100k_Dataset.csv with both the customer and product lookup
 DataFrames.
# Temporarily reset the index of df to make OrderID a column for merging
df_reset = df.reset_index().copy()

# Merge with customer_lookup_df to add 'TotalOrders'
# We only need 'CustomerID' and 'TotalOrders' from the lookup table for this merge
df_merged_customer = pd.merge(df_reset, customer_lookup_df[['CustomerID', 'TotalOrders']],
                              on='CustomerID', how='left')

# Merge the result with product_lookup_df to add 'AveragePrice'
# Ensure Product and Category names match after previous normalization steps
df_merged_final = pd.merge(df_merged_customer, product_lookup_df,
                           on=['Product', 'Category'], how='left')

# Set OrderID back as the index
df = df_merged_final.set_index('OrderID')

print("DataFrame after merging with customer and product lookup DataFrames (first 5 rows):")
display(df.head())
```
